```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/mac/Desktop/DoneDeal WebScraping/DoneDealCars.csv' )

# Calculate median prices by car brand
brand_median_prices = df.groupby('CarBrand')['Price'].median().sort_values(ascending=False)

# Select top 10 brands with highest median prices
top_brands = brand_median_prices.head(10)

# Filter the DataFrame to include only the selected brands
filtered_df = df[df['CarBrand'].isin(top_brands.index)]

# Create a box plot grouped by car brand
plt.figure(figsize=(10, 6))
filtered_df.boxplot(column='Price', by='CarBrand', vert=False)

# Customize the plot
plt.title('Car Prices by Brand (Top 10)')
plt.xlabel('Price')
plt.ylabel('Car Brand')

# Display the plot
plt.show()

#In this code, brand_median_prices calculates the median prices for each car brand and sorts them in descending 
#order. The top_brands variable selects the top 10 brands with the highest median prices. 
#Then, the DataFrame is filtered using isin() to include only the selected brands. 
#Finally, the box plot is created with the filtered data, and the plot title is updated to reflect the top 10 brands being displayed.


#1. Sort the car brands by median price: This will help identify the brands with higher or lower median prices.
#2. Select a subset of car brands to display: Instead of plotting all 50 brands, you can choose to display a subset 
#of brands based on certain criteria, such as the top 10 brands with the most listings or the brands with the highest median prices.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/mac/Desktop/DoneDeal WebScraping/DoneDealCars.csv' )

brand_counts = df['CarBrand'].value_counts()

# Select top 10 brands with the most listings
top_brands = brand_counts.head(10)

# Plot the count of listings using a bar plot
plt.figure(figsize=(10, 6))
top_brands.plot(kind='bar')
plt.title('Top 10 Car Brands with the Most Listings')
plt.xlabel('Car Brand')
plt.ylabel('Count of Listings')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/mac/Desktop/DoneDeal WebScraping/DoneDealCars.csv' )

# Create a scatter plot of miles vs. price for the top 10 brands
top_brands = df['CarBrand'].value_counts().head(10)
filtered_df = df[df['CarBrand'].isin(top_brands.index)]

plt.figure(figsize=(10, 6))
for brand in top_brands.index:
    brand_df = filtered_df[filtered_df['CarBrand'] == brand]
    sample_indices = brand_df.sample(frac=0.5).index
    plt.scatter(brand_df.loc[sample_indices, 'Miles'].astype(str), brand_df.loc[sample_indices, 'Price'], label=brand)

plt.title('Miles vs. Price for Top 10 Car Brands')
plt.xlabel('Miles')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/Users/mac/Desktop/DoneDeal WebScraping/DoneDealCars.csv' )

# Define custom bins with a logarithmic scale
bins = np.logspace(np.log10(100), np.log10(df['Price'].max()), num=20)

# Create a histogram of car prices with custom bins
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=bins)
plt.xscale('log')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
